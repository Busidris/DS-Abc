```python
def remove_substring(name):
    start = name.find("(")
    end = name.find(")")
    new_name = name[0:start] + name[end+1:]
    new_name = new_name.strip()
    return new_name 


```

```python
df['Name_trimmed'] = df['Name'].apply(remove_substring) 
```
